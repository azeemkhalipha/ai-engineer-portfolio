```python
import os
import time
import json
from datetime import datetime, timezone
from groq import Groq
```

```python
def call_groq_and_log(prompt: str, log_filename: str = os.path.expanduser("~/genailearning/call_log.jsonl")):
    """
    Sends a prompt to the Groq API and logs usage metrics to a JSONL file.
    """
    # 1. Setup client and pricing
    client = Groq(
        # This is the default and can be omitted, but included for clarity
        api_key=os.environ["GROQ_API_KEY"]
    )
    
    model_name = "llama-3.1-8b-instant"
    cost_per_input_token = 0.05 / 1_000_000
    cost_per_output_token = 0.08 / 1_000_000

    # 2. Record start time for latency tracking
    start_time = time.perf_counter()
    try:
        # 3. Make the API call
        response = client.chat.completions.create(
            messages=[
                {
                    "role": "user",
                    "content": prompt,
                }
            ],
            model=model_name,
        )

        # 4. Calculate latency in milliseconds
        end_time = time.perf_counter()
        latency_ms = (end_time - start_time) * 1000

        # 5. Extract token usage and calculate total cost
        prompt_tokens = response.usage.prompt_tokens
        completion_tokens = response.usage.completion_tokens
        
        total_cost_usd = (prompt_tokens * cost_per_input_token) + (completion_tokens * cost_per_output_token)

        # 6. Build the JSON log entry
        log_entry = {
            "timestamp": datetime.now(timezone.utc).isoformat(),
            "model": model_name,
            "prompt_tokens": prompt_tokens,
            "completion_tokens": completion_tokens,
            "total_cost_usd": total_cost_usd,
            "latency_ms": round(latency_ms, 2)
        }

        # 7. Append to the JSONL file
        with open(log_filename, "a", encoding="utf-8") as log_file:
            log_file.write(json.dumps(log_entry) + "\n")
        return response.choices[0].message.content

    except Exception as e:
        latency_ms = (time.perf_counter() - start_time) * 1000
        error_entry = {
            "timestamp": datetime.now(timezone.utc).isoformat(),
            "model": model_name,
            "error": str(e),
            "prompt_tokens": None,
            "completion_tokens": None,
            "total_cost_usd": None,
            "latency_ms": round(latency_ms, 2)
        }
        with open(log_filename, "a", encoding="utf-8") as f:
            f.write(json.dumps(error_entry) + "\n")
        raise
```

```python
if __name__ == "__main__":
    sample_prompt = "Explain why the sky is blue in exactly two sentences."
    
    print(f"Sending prompt: '{sample_prompt}'...\n")
    
    try:
        reply = call_groq_and_log(sample_prompt)
        print(f"Response:\n{reply}\n")
        print("Metrics successfully logged to call_log.jsonl.")
    except Exception as e:
        print(f"An error occurred: {e}")
```
